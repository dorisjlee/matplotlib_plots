```python
import matplotlib.pyplot as pyplt
import numpy as np
from scipy import integrate
```

```python
L = 10
```

```python
arr = np.arange(101)
```

```python
arr 
```

```python
#Calculating An
#let f(x) =1 for now.
#The 2nd value in the returned tuple is error in numerical integration (which we will ignore for now)
def a(n):
    return scipy.integrate.quad(lambda x: cos(n*pi*x/L), -L, L) [0] /L
def b(n):
    return scipy.integrate.quad(lambda x: sin(n*pi*x/L), -L, L) [0] /L
```

```python
#Doing the first 100 partial sums 
for x in arr:
    print (a(x))
    print (b(x))
```

```python
def f(x):
    sum =0
    for n in arr: 
        sum = sum + a(n)*cos(n*pi*x/L) + b(n)*sin(n*pi*x/L)
    return a(0)/2 +sum
```
